Load pre-trained gpt2-large Model

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pre-trained GPT-2 tokenizer and model
model_name = "gpt2-large"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Set model to evaluation mode
model.eval()
```

Define Text Generation Function

```python
def generate_text(prompt, max_length=150):
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    with torch.no_grad():
        outputs = model.generate(
            inputs,
            max_length=max_length,
            num_return_sequences=1,
            no_repeat_ngram_size=3,
            do_sample=True,
            top_k=40,
            top_p=0.92,
            temperature=0.7,
            pad_token_id=tokenizer.eos_token_id
        )

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text
```

```python
prompt = "Kabilan is the boy who always"
generated = generate_text(prompt)
```

Output

```python
print(generated)
```

# News Artical Generation

```python
from transformers import pipeline, set_seed

# Load generator pipeline
generator = pipeline("text-generation", model="EleutherAI/gpt-neo-1.3B")

set_seed(42)
def generate_news_article(title):
    prompt = f"Title: {title}\n\nArticle:"
    result = generator(
        prompt,
        max_length=250,
        do_sample=True,
        temperature=0.7,
        top_k=40,
        top_p=0.92,
        no_repeat_ngram_size=3
    )[0]['generated_text']
    return result
```

```python
title = "Scientists discover water on Mars"
article = generate_news_article(title)
print(article)
```
